# Lego Sales Data Cleaning

Exploratory Data Analysis being performed on the data provided by lego.

Data was taken from github

```python
#importing pandas library
import pandas as pd
```

Reading all the necessary datasets to be used to perform Analysis.

```python
#reading all the necessary datasets using pandas 
df=pd.read_csv("https://raw.githubusercontent.com/KeithGalli/lego-analysis/master/datasets/lego_sets.csv")
theme=pd.read_csv("https://raw.githubusercontent.com/KeithGalli/lego-analysis/master/datasets/parent_themes.csv")
df.head(100)
```

Q1. what percentage of all the albums released were star wars themed?

```python
merged[merged['set_num'].isnull()].shape
merged.head()
```

```python
merged=df.merge(theme,left_on="parent_theme",right_on="name")
merged.drop(columns="name_y",inplace= True)

licensed=merged[merged['is_licensed']]
licensed=licensed.dropna(subset='set_num')

licensed.isnull()
licensed.head()


```

```python
starwars_force=licensed[licensed["parent_theme"]=="Star Wars"]
starwars_force.head(10)
```

```python
Force=int(starwars_force.shape[0]/licensed.shape[0]*100)
print(Force)
```

# Here we get the insight that 51% of the albums released were star wars themed.

Q2.In which year was star wars not the most licensed album.


```python

licensed_sort=licensed.sort_values('year')
licensed_sort['count']=1
sumed=licensed_sort.groupby(['year','parent_theme']).sum().reset_index()
sumed.head()
```

```python
sumed_max=sumed.sort_values('count',ascending=False).drop_duplicates(['year'])
sumed_max.head()
```

# we can see that in year 2017 star wars was not the most sold album
